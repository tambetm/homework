```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
envs = [
    ('Ant-v2', 4810.845720136646, 132.78104615579062),
    ('HalfCheetah-v2', 4130.340914316427, 79.42298677938922),
    ('Hopper-v2', 3778.4560274944365, 4.297159652223931),
    ('Humanoid-v2', 10421.483716949297, 62.42961081207094),
    ('Reacher-v2', -3.522377615811832, 1.316911971668481),
    ('Walker2d-v2', 5535.22325118935, 51.97314462289373)
]
```

```python
plt.figure(figsize=(16, 8))
for i, (env, mean, std) in enumerate(envs):
    plt.subplot(2, 3, i + 1)
    with open("bc_results/%s.csv" % env, "rb") as f:
        data = np.loadtxt(f, delimiter=",", skiprows=1)
    plt.plot(data[:, 0], data[:, 1])
    plt.fill_between(data[:, 0], data[:, 1] - data[:, 2], data[:, 1] + data[:, 2], alpha=0.1)

    with open("dagger_results/%s.csv" % env, "rb") as f:
        data = np.loadtxt(f, delimiter=",", skiprows=1)
    plt.plot(data[:, 0], data[:, 1])
    plt.fill_between(data[:, 0], data[:, 1] - data[:, 2], data[:, 1] + data[:, 2], alpha=0.1)

    plt.plot([1, 10], [mean, mean])
    plt.fill_between([1, 10], [mean - std, mean - std], [mean + std, mean + std], alpha=0.1)
    
    plt.legend(["BC", "DAgger", "Expert"])
    plt.title(env)
```
